# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    
    def __init__(self, tv_durumu = "Kapalı", tv_ses = 0, kanal_listesi = ["TRT"], kanal = "TRT"):
        self.tv_durumu = tv_durumu
        self.tv_ses = tv_ses
        self.kanal_listesi = kanal_listesi
        self.kanal = kanal
    
    def tv_ac(self):
        if self.tv_durumu == "Açık":
            print("Televizyon zaten açık.")
        else:
            self.tv_durumu = "Açık"
            print("Televizyon açılıyor...")
            time.sleep(1)
            print("Televizyon açıldı.")
    
    def tv_kapat(self):
        if self.tv_durumu == "Kapalı":
            print("Televizyon zaten kapalı.")
        else:
            self.tv_durumu = "Kapalı"
            print("Televizyon kapatılıyor...")
            time.sleep(1)
            print("Televizyon kapandı.")
    
    def ses_ayarları(self):
        while True:
            cevap = input("Sesi Azalt: '<'\nSesi Artır: '>'\nÇıkış : çıkış")
            if cevap == "<":
                if self.tv_ses != 0:
                    self.tv_ses -= 1
                    print("Ses:", self.tv_ses)
            elif cevap == ">":
                if self.tv_ses != 31:
                    self.tv_ses += 1
                    print("Ses:", self.tv_ses)
            else:
                print("Ses güncellendi:", self.tv_ses)
                break
    
    def kanal_ekle(self, kanal_ismi):
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal_ismi)
        print("Kanal eklendi.")
    
    def kanal_sayısı(self):
        print("Toplam", len(self.kanal_listesi), "kanal var.")
    
    def rastgele_kanal(self):
        rastgele = random.randint(0, len(self.kanal_listesi)-1)
        self.kanal = self.kanal_listesi[rastgele]
        print("Şu anki kanal:", self.kanal)
    
    def televizyon_bilgileri(self):
        print("Tv Durumu:", self.tv_durumu)
        print("Tv Ses:", self.tv_ses)
        print("Kanal Listesi:", self.kanal_listesi)
        print("Şu anki kanal:", self.kanal)

kumanda = Kumanda()

while True:
    print("""
    Televizyon Kumandası
    
    1. Tv Aç
    2. Tv Kapat
    3. Ses Ayarları
    4. Kanal Ekle
    5. Kanal Sayısını Öğrenme
    6. Rastgele Kanala Geçme
    7. Televizyon Bilgileri
    
    Çıkmak için q'ya basın.
    """)
    
    işlem = input("İşlemi Seçiniz:")
    
    if işlem == "q":
        print("Program Sonlandırılıyor...")
        break
    
    elif işlem == "1":
        kumanda.tv_ac()
    
    elif işlem == "2":
        kumanda.tv_kapat()
    
    elif işlem == "3":
        kumanda.ses_ayarları()
    
    elif işlem == "4":
        kanal_isimleri = input("Kanal isimlerini ',' ile ayırarak girin:")
        kanal_listesi = kanal_isimleri.split(",")
        for kanal in kanal_listesi:
            kumanda.kanal_ekle(kanal)
    
    elif işlem == "5":
        kumanda.kanal_sayısı()
    
    elif işlem == "6":
        kumanda.rastgele_kanal()
    
    elif işlem == "7":
        kumanda.televizyon_bilgileri()
    
    else:
        print("Geçersiz İşlem") 
```
